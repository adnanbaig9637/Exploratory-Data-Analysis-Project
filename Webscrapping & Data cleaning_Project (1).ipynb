```python
import requests as req
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re
```

```python
url1 = "https://www.cars24.com/buy-used-car?sort=lhl&serveWarrantyCount=true&listingSource=Homepage_Filters&storeCityId=3686&pinId=500001"
resp = req.get(url1)
resp.status_code
```

```python
soup = BeautifulSoup(resp.text)
soup.text
```

```python
Year=[]
car_name=[]
Add = []
Kmt=[]
fuel=[]
Trans=[]
Price = []
kmt1 =[]
for i in soup.find_all('div',class_="_2z-Yu"):
    year = i.find('h3',class_='_2lmIw').text
    pattern= r"\b\d{4}\b"
    regex= re.findall(pattern,year)
    if regex:
        Year.append(regex[0])
    else:
        Year.append(np.nan)

for i in soup.find_all('div',class_="_2z-Yu"):
    year = i.find('h3',class_='_2lmIw').text
    pattern= " \w+"
    regex= re.findall(pattern,year)
    if regex:
        car_name.append(regex)
    else:
        car_name.append(np.nan)

for i in soup.find_all('div',class_="_2z-Yu"):
    add = i.find('span',class_="_3DYbK")
    if add:
        Add.append(add.text)
    else:
        Add.append(np.nan)
Kmt=[]
for i in soup.find_all('div',class_="_2z-Yu"):
    kmt = i.find('ul',class_='_13yb6').text
    pattern= r"\d{1,3},\d{2,3}|[,]\d{1,3}"
    regex= re.findall(pattern,kmt)
    if regex:
        Kmt.append(regex)
    else:
        Kmt.append(np.nan)

fuel=[]
for i in soup.find_all('div',class_="_2z-Yu"):
    kmt = i.find('ul',class_='_13yb6').text
    pattern= r"Petrol|Diesel"
    regex= re.findall(pattern,kmt)
    if regex:
        fuel.append(regex)
    else:
        fuel.append(np.nan)



for i in soup.find_all('div',class_="_2z-Yu"):
    trans = i.find('ul',class_='_1hOnS').text
    pattern= r"Manual|Automatic"
    regex= re.findall(pattern,trans)
    if regex:
        Trans.append(regex)
    else:
        Trans.append(np.nan)


for i in soup.find_all('div',class_="_2z-Yu"):
    price= i.find('div',class_="_18ToE").text
    yoo = price.split("₹")[1].replace(",","")
    Price.append(yoo)
```

```python
print(Year)
print(car_name)
print(Add)
print(Kmt)
print(fuel)
print(Trans)
print(Price)
```

```python
dicto = {'Name of Cars':car_name,'Registration Year':Year,'Kmt run':Kmt,'Transmission':Trans,'Fuel type':fuel,'Price':Price,'Location of car':Add}
```

```python
kmt1 = []
dicto["Kmt run"]
for i in dicto["Kmt run"]:
    a = str(i).replace('[','')
    b = a.replace(' ','')
    c = b.replace(',','')
    d = c.replace(']','')
    e = d.replace("''",'')
    kmt1.append(int(e.strip("'")))
    
```

```python
dicton = {'Name of Cars':car_name,'Registration Year':Year,'Kmt run':kmt1,'Transmission':Trans,'Fuel type':fuel,'Price':Price,'Location of car':Add}
dicton
```
